```python
import pandas as pd
```

```python
data = pd.read_csv('heart.csv')
```

```python
data.head()
```

```python
bins = [0, 35, 50, 65, 100]
labels = ['Young', 'Middle-aged', 'Senior', 'Elderly']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels)
```

```python
data.head()
```

```python
bp_bins = [0, 120, 140, 200]
bp_labels = ['Normal', 'Elevated', 'High']
data['bp_level'] = pd.cut(data['trestbps'], bins=bp_bins, labels=bp_labels)
```

```python
data.head()
```

```python
data['high_chol'] = data['chol'] > 240
data['high_bp'] = data['trestbps'] > 140
data['high_fbs'] = data['fbs'] > 120
```

```python
data['risk_factors'] = data[['high_chol', 'high_bp', 'high_fbs']].sum(axis=1)
```

```python
data.head()
```

```python
data.to_csv('feature_engineered.csv', index=False)
```
