```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
iris = sns.load_dataset('iris')
```

Exercise 1

```python
sepal_length_mean = iris['sepal_length'].mean()
sepal_length_median = iris['sepal_length'].median()
sepal_length_mode = iris['sepal_length'].mode()[0]
```

```python
print(f"Sepal Length - Mean: {sepal_length_mean}, Median: {sepal_length_median}, Mode: {sepal_length_mode}")
```

Exercise 2

```python
petal_width_variance = iris['petal_width'].var()
petal_width_std_dev = iris['petal_width'].std()
```

```python
print(f"Petal Width - Variance: {petal_width_variance}, Standard Deviation: {petal_width_std_dev}")
```

Exercise 3

```python
summary_table = iris.drop(columns=['species']).describe().T
summary_table['median'] = iris.drop(columns=['species']).median()
summary_table['variance'] = iris.drop(columns=['species']).var()
summary_table = summary_table[['mean', 'median', 'variance', 'std']]
print("Summary Table:\n", summary_table)
```

Exercise 4

```python
sepal_length_counts = iris['sepal_length'].value_counts(normalize=True).sort_index()
```

```python
print("Probability Distribution of Sepal Lengths:\n", sepal_length_counts)
```

Exercise 5

```python
sns.histplot(iris['sepal_length'], kde=True)
plt.title('Probability Distribution of Sepal Lengths')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

Exercise 6

```python
petal_length_sorted = np.sort(iris['petal_length'])
petal_length_cdf = np.arange(1, len(petal_length_sorted) + 1) / len(petal_length_sorted)
```

```python
plt.plot(petal_length_sorted, petal_length_cdf, marker='.', linestyle='none')
plt.title('Cumulative Distribution Function (CDF) of Petal Lengths')
plt.xlabel('Petal Length')
plt.ylabel('CDF')
plt.show()
```

Exercise 7

```python
sns.kdeplot(iris['sepal_width'])
plt.title('Probability Density Function (PDF) of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Density')
plt.show()
```

Exercise 8

```python
given_value = 4.5
prob_greater = (iris['petal_length'] > given_value).mean()
```

```python
print(f"Probability of a petal length greater than {given_value}: {prob_greater}")
```

Exercise 9

```python
species1 = 'setosa'
species2 = 'versicolor'
t_stat, p_value = stats.ttest_ind(iris[iris['species'] == species1]['petal_length'],
                                  iris[iris['species'] == species2]['petal_length'])
```

```python
print(f"Hypothesis Test - t-statistic: {t_stat}, p-value: {p_value}")
```

Exercise 10

```python
covariance = np.cov(iris['sepal_length'], iris['sepal_width'])[0, 1]
correlation = iris['sepal_length'].corr(iris['sepal_width'])
```

```python
print(f"Covariance between sepal length and sepal width: {covariance}")
print(f"Correlation between sepal length and sepal width: {correlation}")
```
