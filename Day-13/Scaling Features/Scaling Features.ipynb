```python
import pandas as pd
```

```python
data = pd.read_csv('WineQT.csv')
```

```python
data.head()
```

```python
print(data.describe())
```

Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_n = MinMaxScaler()
```

```python
features = data.drop(columns=['quality', 'Id'])
```

```python
normalized = scaler_n.fit_transform(features)
```

```python
normalized_df = pd.DataFrame(normalized, columns=features.columns)
```

```python
print(normalized_df.describe())
```

Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
standardized = scaler.fit_transform(features)
```

```python
standardized_df = pd.DataFrame(standardized, columns=features.columns)
```

Analyze the Effects

```python
import matplotlib.pyplot as plt
```

```python
data.drop(columns=['quality', 'Id']).hist(figsize=(15, 10))
plt.suptitle('Histograms of Original Features')
plt.show()
```

```python
normalized_df.hist(figsize=(15, 10))
plt.suptitle('Histograms of Normalized Features')
plt.show()
```

```python
standardized_df.hist(figsize=(15, 10))
plt.suptitle('Histograms of Standardized Features')
plt.show()
```
